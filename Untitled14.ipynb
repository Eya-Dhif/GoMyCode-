```python
import requests
from bs4 import BeautifulSoup

#get html content
def get_parse(url):
  ur= requests.get(url)
  return BeautifulSoup(ur.content , 'html.parser')
url= 'https://en.wikipedia.org/wiki/Arabic'

def wikibot (url):
   url_open= requests.get(url)
   soup= BeautifulSoup(url_open.content , 'html.parser')
   details = soup ('table',{'class':'infobox'})
   for i in details :
       h = i.find_all ('tr')
       for j in h :
         heading = j.find_all('th')
         detail = j.find_all ('td')
         if heading is not None and detail is not None :
           for x,y in zip(heading,detail) :
             print ("{}  ::  {}".format(x.text,y.text))
             print("--------------------------------------------------------------")
   for i in range (1,3) :
        print(soup('p')[i].text)
wikibot(url)

#collect every link that redirects to another Wikipedia page
def collect_all_links(soup):
  links = soup.find_all('a')
  return [link.get('href') for link in links]

#Wrap all the previous functions into a single function
def wrap(url):
    soup = get_parse(url)
    wikibot(url)
    links = collect_all_links(soup)
    for link in links:
        print(link)

#test
wrap('https://en.wikipedia.org/wiki/Arabic')





```
