```python
import pandas as pd
import numpy as np

exam_data = {
      'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],

      'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],

      'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],

      'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df= pd.DataFrame(exam_data, index=labels)
print(df)
df.info()

#Print the three first rows using the head() method
df.head(3)

#Delete rows with Nan values
without_nan= df.dropna()
print(without_nan)

#Extract the 'name' and 'score' columns from the DataFrame
name_score= df[["name", "score"]]
print(name_score)

#append a new row 'k' to the DataFrame
row_k= pd.DataFrame({"name": "Suresh", "score": 15.5, "attempts": 1, "qualify": "yes"}, index=['k'])
new_data= pd.concat([df, row_k])
print(new_data)

#delete the 'attempts' column from the DataFrame
without_attempts= df.drop("attempts", axis=1)
print(without_attempts)

#function to detremine "success"
def determine_success(score):
  if score >10:
    return 1
  else:
    return 0

#Add a new column "Success": if the score is higher than 10 we will have 1, else we will have 0
df["success"]=df["score"].apply(determine_success)
print(df)

#export it into a CSV file named "my_data"
my_data= pd.DataFrame(exam_data)
my_data.to_csv=("my_data.csv")
```
